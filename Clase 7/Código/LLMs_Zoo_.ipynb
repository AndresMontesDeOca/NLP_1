Generación de Texto con GPT

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
model_name = "gpt2"
model = AutoModelForCausalLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

Pad token ya que gpt 2 no lo tiene.

```python
tokenizer.pad_token = tokenizer.eos_token
```

Generacion de texto

```python
prompt = "The future of artificial intelligence will"
inputs = tokenizer(prompt, return_tensors="pt")
```

```python
outputs = model.generate(
    inputs["input_ids"],
    max_length=50,
    num_return_sequences=3,
    temperature=0.8,
    do_sample=True,
    pad_token_id=tokenizer.eos_token_id
)
```

```python
for i, output in enumerate(outputs):
    generated_text = tokenizer.decode(output, skip_special_tokens=True)
    print(f"Resultado {i+1}: {generated_text}")
```

Traducción con T5

```python
from transformers import T5ForConditionalGeneration, T5Tokenizer
```

```python
t5_model_name = "t5-small"
t5_tokenizer = T5Tokenizer.from_pretrained(t5_model_name)
t5_model = T5ForConditionalGeneration.from_pretrained(t5_model_name)
```

Funcion para traduccion

```python
def translate_text(text, source_lang="en", target_lang="es"):
    input_text = f"translate {source_lang} to {target_lang}: {text}"
    inputs = t5_tokenizer(input_text, return_tensors="pt", max_length=512, truncation=True)

    outputs = t5_model.generate(
        inputs["input_ids"],
        max_length=150,
        num_beams=4,
        early_stopping=True
    )

    translation = t5_tokenizer.decode(outputs[0], skip_special_tokens=True)
    return translation
```

Traduccion

```python
texts_to_translate = [
    "Hello, how are you today?",
    "Machine learning is a subset of artificial intelligence.",
    "The weather is beautiful today.",
    "I love programming and technology."
]

for text in texts_to_translate:
    translation = translate_text(text, "en", "es")
    print(f"Inglés: {text}")
    print(f"Español: {translation}")
```

Resumen con T5

Funcion de resumen de texto

```python
def summarize_text(text, max_length=100):
    input_text = f"summarize: {text}"
    inputs = t5_tokenizer(input_text, return_tensors="pt", max_length=512, truncation=True)

    outputs = t5_model.generate(
        inputs["input_ids"],
        max_length=max_length,
        num_beams=4,
        early_stopping=True,
        length_penalty=2.0
    )

    summary = t5_tokenizer.decode(outputs[0], skip_special_tokens=True)
    return summary
```

```python
long_texts = [
    """
    Artificial intelligence (AI) is intelligence demonstrated by machines, in contrast to the natural
    intelligence displayed by humans and animals. Leading AI textbooks define the field as the study
    of "intelligent agents": any device that perceives its environment and takes actions that maximize
    its chance of successfully achieving its goals. Colloquially, the term "artificial intelligence"
    is often used to describe machines that mimic "cognitive" functions that humans associate with the
    human mind, such as "learning" and "problem solving". As machines become increasingly capable,
    tasks considered to require "intelligence" are often removed from the definition of AI, a phenomenon
    known as the AI effect.
    """
]

for i, text in enumerate(long_texts, 1):
    summary = summarize_text(text.strip(), max_length=80)
    print(f"Texto {i} Original ({len(text.split())} palabras):")
    print(text.strip()[:200] + "...")
    print(f"\nResumen ({len(summary.split())} palabras):")
    print(summary)
```

Generación de Texto con GPT-Neo

```python
gpt_neo_model_name = "EleutherAI/gpt-neo-125m"
gpt_neo_tokenizer = AutoTokenizer.from_pretrained(gpt_neo_model_name)
gpt_neo_model = AutoModelForCausalLM.from_pretrained(gpt_neo_model_name)
```

```python
gpt_neo_tokenizer.pad_token = gpt_neo_tokenizer.eos_token
```

```python
prompt_neo = "In a world where technology dominates,"
inputs_neo = gpt_neo_tokenizer(prompt_neo, return_tensors="pt")
```

```python
outputs_neo = gpt_neo_model.generate(
    inputs_neo["input_ids"],
    max_length=70,
    num_return_sequences=3,
    temperature=0.9,
    do_sample=True,
    top_p=0.9,
    pad_token_id=gpt_neo_tokenizer.eos_token_id
)
```

```python
for i, output in enumerate(outputs_neo):
    generated_text = gpt_neo_tokenizer.decode(output, skip_special_tokens=True)
    print(f"Resultado {i+1}: {generated_text}")
```
