### **Parte A) Tokenization**

#### **Ejercicio I: Crear un Tokenizador Básico**

Descripción: Escribe una función que tome una cadena de texto y la divida en palabras individuales (tokens)

```python
def tokenizar(texto):
    texto = texto.lower()

    signos_puntuacion = ['.', ',', '!', '?', ':', ';', '(', ')', '[', ']', '{', '}', '¿', '¡', '"', "'"]

    for signo in signos_puntuacion:
        texto = texto.replace(signo, ' ')

    tokens = texto.split()

    return tokens

texto = "¡Hola, Mundo! Este es un ejemplo de tokenización"
tokens = tokenizar(texto)
print(tokens)
```

#### **Ejercicio II: Crear un Tokenizador que maneje palabras compuestas**

Descripción: Escribe una función que tome una cadena de texto y la divida en palabras individuales (tokens) pero que no divida palabras compuestas, es decir que la palabra "New York" se mantenga como una sola palabra y no como "New", "York".

Ejemplo:
Entrada: "Mañana viajare a New York"
Salida: "mañana", "viajare", "a", "New York"

```python
def tokenizer(sentence, compound_words):

    sentence = sentence.lower()
    tokens = []

    for compound in compound_words:
        sentence = sentence.replace(compound.lower(), compound.replace(" ", "_"))

    words = sentence.split()

    for word in words:
        if "_" in word:
            tokens.append(word.replace("_", " "))
        else:
            tokens.append(word)

    return tokens


sentence = "Mañana viajare a New York"
compound_words = ["New York", "Los Angeles"]
tokens = tokenizer(sentence, compound_words)
print(tokens)
```

### **Parte B) Preprocesamiento de texto**

```python
# Descargas iniciales necesarias
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')

from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
from nltk.tokenize import word_tokenize, sent_tokenize

# Configuración para español
stop_words_es = set(stopwords.words('spanish'))
stemmer_es = SnowballStemmer('spanish')
```

```python
def imprimir_separador(titulo):
    print(f"\n{'='*20} {titulo} {'='*20}")
```

```python
# --- EJEMPLO 1: TOKENIZACIÓN (EL MICROSCOPIO) ---
imprimir_separador("1. TOKENIZACIÓN")
frase1 = "¡Hola! ¿Cómo estás? El Dr. García llegará a las 5 p.m."
tokens = word_tokenize(frase1)
oraciones = sent_tokenize(frase1)

print(f"FRASE ORIGINAL: {frase1}")
print(f"TOKENS (Palabras y signos): {tokens}")
print(f"ORACIONES DETECTADAS: {oraciones}")
print("> Nota: Observen cómo detectó que 'Dr.' no termina la oración.")
```

```python
# --- EJEMPLO 2: STOPWORDS (FILTRANDO RUIDO) ---
imprimir_separador("2. STOPWORDS (PALABRAS VACÍAS)")
frase2 = "La casa de papel es una serie de televisión española."
palabras = word_tokenize(frase2.lower()) # Pasamos a minúsculas para comparar mejor

# Filtramos: solo dejamos la palabra si NO está en la lista de stopwords

filtrado = [w for w in palabras if w not in stop_words_es and w.isalpha()]
# stop_words_es: Conjunto de palabras frecuentes (artículos, preposiciones) que no aportan significado clave y se eliminan para reducir ruido.
# isalpha(): Método de Python que filtra el texto para conservar solo palabras, descartando números, espacios y signos de puntuación.

print(f"FRASE ORIGINAL: {frase2}")
print(f"TEXTO FILTRADO: {filtrado}")
print(f"> Se eliminaron {len(palabras) - len(filtrado)} palabras de relleno.")
```

```python
frase2_1= "No estoy tan solo, pero no es lo mismo."
palabras = word_tokenize(frase2_1.lower())
filtrado = [w for w in palabras if w not in stop_words_es and w.isalpha()]
print(f"FRASE ORIGINAL: {frase2_1}")
print(f"TEXTO FILTRADO: {filtrado}")
print(f"> Se eliminaron {len(palabras) - len(filtrado)} palabras de relleno.")
```

```python
# --- EJEMPLO 3: STEMMING (BUSCANDO LA RAÍZ "PODANDO") ---
imprimir_separador("3. STEMMING (BUSCANDO LA RAÍZ PODANDO")
frase3 = "Los programadores programan programas en la programación."
tokens3 = word_tokenize(frase3.lower())

raices = [stemmer_es.stem(w) for w in tokens3 if w.isalpha()]

print(f"FRASE ORIGINAL: {frase3}")
print(f"RAÍCES DETECTADAS: {raices}")
```

```python
# --- EJEMPLO 4: LEMATIZACIÓN (BUSCANDO LA SEMILLA) ---
import spacy
# Asegúrate de haber descargado el modelo: !python -m spacy download es_core_news_sm
nlp = spacy.load("es_core_news_sm")

imprimir_separador("4. LEMATIZACIÓN (BUSCANDO LA SEMILLA)")
frase4 = "Los programadores programan programas en la programación."
doc = nlp(frase4.lower())

# Extraemos el lema de cada palabra
lemas = [token.lemma_ for token in doc if token.is_alpha]

print(f"FRASE ORIGINAL: {frase4}")
print(f"LEMAS DETECTADOS: {lemas}")
```

```python
# --- COMPARACIÓN DIRECTA ---
frase_test = "Los mejores deportistas cuentan sus mejores cuentos mientras van al estadio."

# 1. Stemming (NLTK)
tokens_nltk = word_tokenize(frase_test.lower())
resultado_stem = [stemmer_es.stem(w) for w in tokens_nltk if w.isalpha()]

# 2. Lematización (spaCy)
doc_spacy = nlp(frase_test.lower())
resultado_lema = [token.lemma_ for token in doc_spacy if token.is_alpha]

imprimir_separador("COMPARACIÓN FINAL")
print(f"FRASE: {frase_test}")
print(f"STEMMING: {resultado_stem}")
print(f"LEMATIZACIÓN: {resultado_lema}")
```

```python
# --- EJEMPLO 5: ETIQUETADO GRAMATICAL (POS TAGGING) ---
# Nota: El etiquetado por defecto de NLTK funciona mejor en inglés.
# Usaremos una frase simple para ilustrar el concepto.
imprimir_separador("5. POS TAGGING (CATEGORÍAS)")

frase4 = "The quick brown fox jumps. I wait"
tokens4 = word_tokenize(frase4)
etiquetas = nltk.pos_tag(tokens4)

print(f"FRASE: {frase4}")
print(f"ETIQUETAS: {etiquetas}")
print("> Leyenda: NN = Sustantivo, VBZ = Verbo, JJ = Adjetivo, DT = Determinante, PRP= Pronombre personal.")
```

```python
# --- EJEMPLO 6: DISTRIBUCIÓN DE FRECUENCIA ---
imprimir_separador("6. FRECUENCIA DE PALABRAS")
texto_largo = "el gato corre tras el ratón y el perro corre tras el gato"
tokens5 = word_tokenize(texto_largo.lower())

# FreqDist: Crea un diccionario con el conteo de apariciones de cada palabra en el texto.
frecuencia = nltk.FreqDist(tokens5)

print(f"TEXTO: {texto_largo}")
print("TOP 3 PALABRAS MÁS USADAS:")
for palabra, count in frecuencia.most_common(3):
    print(f"- '{palabra}': {count} veces")
```
